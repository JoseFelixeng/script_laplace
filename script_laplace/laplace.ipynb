```python
# Script para resolver laplace com python

# pacote simbolico
import sympy
sympy.init_printing()

#geração dos simbolos
s,t = sympy.symbols('s t')

```

```python
# Função Transformada de Laplace

def L(f):
	return sympy.laplace_transform(f, t, s, noconds=True)

# Função transformada de Laplace Inversa

def invL(F):
	return sympy.inverse_laplace_transform(F, s, t)
```

```python
# Transformada de Laplace da entrada

F = L(sympy.exp(-2*t)*2)
F
```

```python
# Transformada de Laplace da entrada

G = 36/(s**2 + 4.2*s + 36)
G
```

```python
# A resposta descrita no Domínio de Laplace

X = G * F
X
```

```python
# Expansão em Frações Parciais da resposta
G.apart(s)

```

```python
# A resposta temporal via transformada Inversa de Laplace

invL(X)
```
